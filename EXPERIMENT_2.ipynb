QUESTION);-  MARKS = [90,82,66,76,33]

FIND USING list

1): MAXIMUM MARKS

2): MINIMUM MARKS

3): TOTAL NO OF STUDENT

4): SUM OF MARKS

5): AVERAGE MARKS OF ALL THE STUDENTS

6): SORT IN ASCENDING ORDER

```python
MARKS = [90, 82, 66, 76, 33]

# MAXIMUM MARKS
max_marks = max(MARKS)
print(f"Maximum marks: {max_marks}")

# MINIMUM MARKS
min_marks = min(MARKS)
print(f"Minimum marks: {min_marks}")

# TOTAL NO OF STUDENT
total_students = len(MARKS)
print(f"Total number of students: {total_students}")

# SUM OF MARKS
sum_of_marks = sum(MARKS)
print(f"Sum of marks: {sum_of_marks}")

# AVERAGE MARKS OF ALL THE STUDENTS
average_marks = sum(MARKS) / len(MARKS)
print(f"Average marks: {average_marks:.2f}")

# SORT IN ASCENDING ORDER
sorted_marks = sorted(MARKS)
print(f"Marks sorted in ascending order: {sorted_marks}")
```
